```python
import matplotlib.pyplot as plt
```

```python
plt.hist(df['Age'].dropna(), bins=20)
plt.xlabel('Age')
plt.ylabel('Count')
plt.title('Age Distribution')
plt.show()
```

Majority of the people travelling are between the age groups of 25 - 25. Which means they are young aspirants travelling either for business(Mostly) and a few for personal reasons or pleasure. Therefore it would be wise to come up with deals which can be accomodated for these business travellers. B2B travels are exclusive and they have easier access to airports and all their facilities are premium already, but an addition to these packages, even small gestures such as an additional snack while they travel or a cab service to their destination(if not the whole trip, maybe a discount coupon can help gain more customers for the travel agency).

```python
prod_counts = df['ProdTaken'].value_counts()
plt.bar(prod_counts.index, prod_counts.values)
plt.xticks([0, 1], ['No', 'Yes'])
plt.xlabel('Product Taken')
plt.ylabel('Count')
plt.title('Product Taken by Customers')
plt.show()
```

A lot of people are not availing our facilities to the highest extend, that may be due to the prices or the companies we are in association with, which means, coming up with better relations and facilities provision is of utmost importance so that we can gain more customers and to retain them, we can come up with loyalty programs to get a hold on the market and capture even more market portion.

```python
plt.scatter(df['MonthlyIncome'], df['PitchSatisfactionScore'])
plt.xlabel('Monthly Income')
plt.ylabel('Pitch Satisfaction Score')
plt.title('Monthly Income vs Pitch Satisfaction')
plt.show()
```

The satisfaction level lies on all stages and we barely see any outliers, but for the one's on a higher scale, they are not very satisfied by the services provided. Therefore, we can conclude that we are lacking behind on the services part. We must employ a customer satisfaction survey and a feedback system to gain insights as to where we are lacking and work on them immidiately. The services ranges on on 1 to 5 scale and therefore, we must work on the one's with lower ratings so that we do not lose onto those customers. A feedback system and a loyalty program with exclusive benefits can be a good way to tackle this issue.

```python
city_tier_counts = df.groupby('CityTier').size()
plt.bar(city_tier_counts.index, city_tier_counts.values)
plt.xlabel('City Tier')
plt.ylabel('Number of Customers')
plt.title('Count of Customers by City Tier')
plt.show()
```

People are targeting lower tier cities but instead we should slowly shift our focus to mid tier cities and get people to take those packages instead of the lower tier cities. Doing so, although would increase the costs and we might lose onto the customers but, the one's who are loyal will definately upgrade and for the one's who still focus on the lower tiers, maybe coming up with better stay inn options or travel facilities can be useful. This way, we can work on retaining the lost customer share in the market. As for the upper tier cities, we must work on providing luxary facilites so that they enjoy their stays even more and come up with package criterias, such as lower tier, mid term and high tier and a 4th one, Luxary Tier packages so that people have flexibility to choose from them and they can enjoy their travels more.
